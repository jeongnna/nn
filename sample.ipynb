```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from pyheaan import nn
```

```python
# 적당한 데이터 가지고 numpy.ndarray 형태로 X, y를 만들면 됩니다.
# X에는 intercept 제외하고 feature들만 넣고, y에는 0 또는 1 값
df = pd.read_csv('santander_50k.csv')
X = df.drop(['ID_code', 'target'], axis=1).values # shape: (N, D)
y = df['target'].values # shape: (N, )
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = nn.models.LogisticRegression(num_features=X_train.shape[1], num_category=2)
```

```python
# model fit 함수에 batch 기능이 없어서 이렇게.. 누가 만들어주세요
for _ in range(100):
    idx = np.random.randint(X_train.shape[0], size=512)
    model.fit(X_train[idx, :], y_train[idx], n_iter=1)
```

```python
pred = model.predict(X_test)
pred = np.argmax(pred, axis=1)
#pred = pred > 0.5
```

```python
accuracy = sum(y_test == pred) / len(y_test)
accuracy # initial: 0.900866
```
